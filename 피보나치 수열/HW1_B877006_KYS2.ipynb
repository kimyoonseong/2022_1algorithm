```python
#hanoi tower             a:시작 b:목표 c:임시
def hanoi(n,a,b,c):
  count=0
  if(n>0):
      #hanoi(n-1,a,c,b)
      count+=hanoi(n-1,a,c,b) #return 값이 count 이므로 계속 카운트에 더해준다 
      count+=1  #그냥 count에 1씩 더하기만 하면 재귀함수로 계속 재호출 되므로 count 누적이 안된다. 
      print(a,"에 있는 원반을",b,"로 옮긴다")
      #hanoi(n-1,c,b,a)
      count+=hanoi(n-1,c,b,a)

  return count
```

```python
hanoi(10,'a','b','c')
```
